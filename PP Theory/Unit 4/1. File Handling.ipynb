```python
File Handling in Python:

Python provides built-in functions to create, read, write, and
append files using the open() function.
```

```python
A file operation takes place in the following order:

1. Open a file
2. Read or write (perform operation)
3. Close the file
```

### 1. File Modes:

```python
Mode	Description
"x" 	Creates a new file, error if file exists.
"r" 	Read (default). Opens file, error if file doesn’t exist.
"w" 	Write. Creates new file or overwrites existing file.
"a" 	Append. Adds data at the end of the file.
"r+"	Read and Write.
"w+"	Write and Read (overwrites file).
"a+"	Append and Read.
```

```python
1. Opening a File
```

```python
file = open("filename", "mode")
```

```python
Creating a file
```

```python
file = open("sample3.txt", "x")
print("File created successfully")
```

```python
f = open("sample.txt", "r")
print("File opened successfully")
f.close()
```

```python
2. Writing to a File and Reading from a File
```

```python
f = open("sample.txt", "w")  # Overwrites if file exists
f.write("Hello, this is Python File Handling!\n")
f.write("Writing second line.\n")
f.close()

# Reading back
f = open("sample.txt", "r")
print(f.read())
f.close()
```

```python
3. Reading the whole File - read()

readline(): Reading line by line
```

```python
f = open("sample.txt", "r")
print(f.readline())    # Reads one line
f.close()
```

```python
readlines(): Reading all lines in a list
```

```python
f = open("sample.txt", "r")
print(f.readlines())   # Reads all lines into a list
f.close()
```

```python
writelines(): Writes multiple lines from a list
```

```python
f = open("sample.txt", "w")  # Overwrites if file exists
f.writelines(["Python is interesting. ", "But students are not showing interest"])
f.close()

# Reading back
f = open("sample.txt", "r")
print(f.read())
f.close()
```

```python
4. Appending to a File in append mode
```

```python
f = open("sample.txt", "a")  # Adds to existing content
f.write("This line is appended.\n")
f.close()

# Reading back
f = open("sample.txt", "r")
print(f.read())
f.close()
```

```python
5. Using with Statement (Best Practice)

It automatically closes the file after use.
```

```python
with open("sample.txt", "a") as f:
    f.write("Appended using WITH statement.\n")
    
with open("sample.txt", "r") as f:
    print(f.read())
```

### 2. File Methods:

```python
Once a file is opened, you can use various methods to interact with it.

Method	Description
f.read()	Reads the file content (all or size bytes).
f.readline()	Reads a single line.
f.readlines()	Reads all lines into a list.
f.write(string)	Writes a string into the file.
f.writelines(list)	Writes a list of strings into the file.
f.close()	Closes the file.

f.tell()	Returns the current position of file pointer.
f.seek(offset)	Moves the file pointer to a specific position.
```

```python
tell() Method: It tells (returns) the current position of the file 
pointer in the file.
```

```python
f = open("sample.txt", "w") # Overwrites if file exists
f.write("Hello Python")
f.close()

f = open("sample.txt", "r")
print(f.read(5))      # Read first 5 characters
print("Current position:", f.tell())  # File pointer position
f.close()
```

```python
seek(offset) Method: Moves the file pointer to a specific
position.

offset: number of bytes to move.
```

```python
f = open("sample.txt", "r")
print(f.read(5))       # Read 5 characters → "Hello"
f.seek(0)              # Move pointer to beginning
print(f.read())        # Read full file again
f.close()
```

```python
f = open("sample.txt", "r")
f.seek(6)              # Move to position 6
print(f.read())        # Reads from position 6 → "Python"
f.close()
```

```python
Combination of tell() and seek()
```

```python
f = open("sample.txt", "r")
print("Initial position:", f.tell())

f.read(5)
print("After reading 5 chars:", f.tell())

f.seek(0)
print("After resetting:", f.tell())

f.close()
```

### 4. Working with CSV files

```python
CSV (Comma-Separated Values) is a popular format for storing 
tabular data.
```

```python
1. Writing to a CSV file
```

```python
import csv
with open("students.csv", "w", newline="") as f:
    wr = csv.writer(f)
    wr.writerow(["Name", "Age", "Marks"])   # Header
    wr.writerow(["Archana", 20, 85])
    wr.writerow(["Raj", 22, 90])
    wr.writerow(["Charitha", 21, 78])
```

```python
2. Reading from a CSV file
```

```python
import csv
with open("students.csv", "r") as f:
    re = csv.reader(f)
    for row in re:
        print(row)
```

### 5. Working with JSON Files

```python
JSON (JavaScript Object Notation) is used for storing structured
data (like dictionaries).
```

```python
a) CSV is best for tabular data (rows/columns).

b) JSON is best for structured/nested data.
```

```python
1. Writing to a JSON file
```

```python
import json

data = {
    "name": "Alice",
    "age": 20,
    "marks": [85, 90, 78]
}

with open("student.json", "w") as f:
    json.dump(data, f) 
```

```python
2. Reading from a JSON file
```

```python
import json

with open("student.json", "r") as f:
    data = json.load(f)

print(data)
```
