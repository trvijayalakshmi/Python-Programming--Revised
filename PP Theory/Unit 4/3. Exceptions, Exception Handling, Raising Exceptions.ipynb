## 2. Exception Handling

```python
For Exception handling, python uses following keywords or statements

try
except
else
finally
```

```python
Try block: Lets you test a block of code for errors.
Except block: Lets you handle the error.
Else block: Lets you execute code when there is no error.
Finally block: Always runs (cleanup code), whether exception occurs
or not.
```

```python
To detect exceptions, write a code that might cause an error 
and enclose it in a try-except block. The try block contains the 
code that could potentially raise an exception.

When an exception is raised inside the try block, the program 
jumps to the corresponding except block. The except block contains
the code to handle the exception gracefully, allowing the program 
to continue running.
```

```python
s = "123"  ## "123"  "abc"
try:
    result = int(s)
except:
    print(f"Error: '{s}' is not a valid integer!")
else:
    print(f"The integer value is: {result}")
finally:
    print("Execution finished.")
```

```python
my_list = [1, 2, 3]
try:
    value = my_list[7]
except:
    print("Index out of range.")
else:
    print(f"Value: {value}")
finally:
    print("Execution finished.")
```

### 3. Raising Exceptions Manually

```python
Example 1: Using raise inside try-except
How to create, raise and handle user defined exceptions?
```

```python
try:
    age = int(input("Enter your age: "))
    print("Your age is:", age)
except:
    print("Age cannot be negative")
```

```python
try:
    age = int(input("Enter your age: "))
    if age < 0:
        raise ValueError
    print("Your age is:", age)
except ValueError:
    print("Age cannot be negative")
```
