```python
What is Recursion?

Recursion is a process in which a function calls itself directly
or indirectly to solve a problem.

Every recursive function has:

i) Base Case → The condition that stops recursion.
ii) Recursive Case → Function calls itself with a smaller/simpler
    input.
```

```python
1. Find GCD (Greatest Common Divisor) using Recursion
```

```python
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print(gcd(48, 18))  
```

```python
2. Write a recursive function to print factorial of a Number
```

```python
def factorial(n):
    if n == 0 or n == 1:   # Base case
        return 1
    else:                  # Recursive case
        return n * factorial(n - 1)

print(factorial(5))  # Output: 120
```

```python
def fact(n):
    return 1 if n==0 else n * fact(n-1)
```

```python
3. Write a recursive function to print fibonocci sequence
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n_terms = 10
print("Fibonacci sequence:")
for i in range(n_terms):
    print(fibonacci(i))
```

```python
4. Write a recursive function to check whether the string is
Palindrome or not
```

```python
def is_palindrome(s):
    if len(s)<=1: return True
    if s[0]!=s[-1]: return False
    return is_palindrome(s[1:-1])
print(is_palindrome("madam"))  
print(is_palindrome("racecar"))
print(is_palindrome("python")) 
```

```python
5. Write a recursive function to find sum of the digits of a given
number.
```

```python
def sum_of_digits(n):
    if n < 10:
        return n
    else:
        return n % 10 + sum_of_digits(n // 10)

# Example usage
print(sum_of_digits(12345))
```

```python
6. Write a recursive function to convert decimal number to binary.
```

```python
def binary(n):
    if n > 1:
        binary(n//2)
    print(n % 2,end = '')
    
n = int(input("enter the decimal number "))
binary(n)
```
