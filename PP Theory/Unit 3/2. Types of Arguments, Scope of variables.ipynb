### 1. Types of Arguments

```python
Python supports different kinds of parameters:
```

```python
a) Positional arguments → matched by order.

b) Keyword arguments → matched by key.

c) Variable length Positional arguments (*args)

d) Variable length Keyword arguments (*kargs)
```

```python
a) Positional arguments (order matters)
```

```python
def multiply(x, y):
    return x * y

print(multiply(3, 4))
```

```python
Positional arguments with default value
```

```python
def multiply(x, y=5):
    return x * y

print(multiply(3))
print(multiply(3,6))
```

```python
Calculate compound interest given principal, rate and time with
defalut values
```

```python
def cal_interest(principal, rate=5, time=3):
    amount = principal * (1 + rate/100) ** time
    interest = amount - principal
    return interest

interest = cal_interest(30000)
print("The compound interest is:", interest)
interest = cal_interest(30000, 10)
print("The compound interest is:", interest)
interest = cal_interest(30000, 10, 4)
print("The compound interest is:", interest)
```

```python
b) Keyword arguments (name=value): Order doesn't matter
```

```python
def student_info(name, age):
    print(f"Name: {name}, Age: {age}")

student_info(age=32, name="Ravi")
```

```python
Keyword arguments with default value
```

```python
def student_info(name, age = 20):
    print(f"Name: {name}, Age: {age}")

student_info(name="Ravi")
student_info(name="Ravi", age = 40)
```

```python
def diff(a,b):
    return a-b

c = diff(20,10)
print(c)
c = diff(b = 10, a = 20)
print(c)
```

```python
c) Variable length positional arguments:
```

```python
Multiple Positional Arguments (*args)
*args is used in a function when you want to pass a variable
number of positional arguments.
```

```python
def add_numbers(*nums):
    total = 0
    for n in nums:
        total += n
    return total

print(add_numbers(10, 20))
print(add_numbers(1, 2, 3, 4, 5))
print(add_numbers())
```

```python
Mixing Normal Arguments and *args
```

```python
def add_numbers(a, *nums):
    total = 0
    for n in nums:
        total += n
    total += a
    return total

print(add_numbers(10))
print(add_numbers(10, 20, 40))
print(add_numbers())  # Try this you will get error
```

```python
# Variable length arguments with default value for fixed variable
def add_numbers(a=20, *nums):
    total = 0
    for n in nums:
        total += n
    total += a
    return total

print(add_numbers(10))
print(add_numbers(10, 20, 40))
print(add_numbers())
```

```python
def display_names(*names):
    for i in names:
        print(i)

display_names("Ravi", "Anu", "Kiran")
```

```python
def student_marks(name, *marks):
    print("Student:", name)
    print("Marks:", marks)

student_marks("Ravi", 85, 90, 88)
```

```python
d) Variable length Keyword Arguments (**kwargs)
**kwargs allows us to pass any number of keyword arguments.
```

```python
def student_info(**details):
    for key, value in details.items():
        print(f"{key}: {value}")

student_info(name="Meena", age=20, course="Python")
student_info()
student_info(name="Radha", course="Java")
```

```python
Mixing Normal Arguments and **kwargs
```

```python
def employee(name, **details):
    print("Name:", name)
    for key, value in details.items():
        print(key, ":", value)

employee("Kiran", salary=40000, department="ECE", city="Bangalore")
employee("Sanjay", salary=80000, department="IT", Institute="MGIT")
employee("Sanjay", department="IT", Institute="MGIT")
# employee() # Try this you will get error
```

### 2. Scope of Variables

```python
a) Local variable → Declared inside a function, can only be
accessed within that function.
```

```python
def my_function():
    x = 10   # local variable
    print("Inside function, x =", x)

my_function()
# print(x)  # ❌ Error: x is local, not accessible outside
```

```python
b) Global variable → Declared outside any function, accessible
anywhere in the program.
```

```python
x = 50   # global variable

def my_function():
    print("Inside function, x =", x)

my_function()
print("Outside function, x =", x)
```

```python
c) Local Variable Shadows Global Variable.
```

```python
x = 20

def my_function():
    x = 99   # local variable with same name
    print("Inside function, x =", x)

my_function()
print("Outside function, x =", x)
```

```python
d) Modifying Global Variable Inside Function.
```

```python
x = 100

def modify():
    global x   # declare x as global
    x = x + 50
    print("Inside function, x =", x)

modify()
print("Outside function, x =", x)
```

```python
e) Nested Functions – nonlocal Keyword.
```

```python
def outer():
    x = 30
    
    def inner():
        nonlocal x
        x = 40
        print("Inside inner:", x)
    
    inner()
    print("Inside outer:", x)

outer()
```
