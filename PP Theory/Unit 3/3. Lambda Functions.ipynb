### 3.  Lambda Functions: 

```python
Normal function: Can have multiple statements and expressions. 
Defined using def with a function name.


Lambda Function: Only one expression allowed. 
Nameless one-line functions using the lambda keyword.
Often used with functions like map(), filter(), sorted()
```

```python
Lambda Syntax: lambda arguments: expression
```

```python
1. Write a lambda function to return the square of a number.
```

```python
square = lambda n: n * n
print(square(5)) 
```

```python
# Conventional way of defining a function
def square(n): return n * n

print(square(5))
```

```python
2. Write a lambda function to return the sum of two numbers.
```

```python
add = lambda a, b: a + b
print(add(3, 7)) 
```

```python
# Conventional function
def add(a,b): 
    return a + b

print(add(3, 7))
```

```python
3. Write a lambda function to return the maximum of two numbers.
```

```python
maximum = lambda a, b: a if a > b else b
print(maximum(12, 8))  
```

```python
def maximum(a,b): return a if a>b else b

print(maximum(12, 8))  
```

```python
4. Write a lambda function to check if a number is even or odd.
```

```python
check_even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(check_even(7))   
```

```python
def check_even(x): return "Even" if x % 2 == 0 else "Odd"

print(check_even(7))   
```

### a. Map Function: Applies a function to each element of an iterable.

```python
Map Syntax:  map(lambda arguments: expression, iterable)
```

```python
Transforms each element of an iterable.
```

```python
1. Create a list of squares for the given list of numbers using map.
```

```python
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, numbers))
print(squares) 
```

```python
Without lambda (Conventional way of defining a function)
```

```python
def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]
print(list(map(square, [1, 2, 3, 4, 5])))
print(list(map(square, numbers)))
```

```python
def square(numbers):
    sq = [i*i for i in numbers]
    return sq

numbers = [1, 2, 3, 4, 5]
print(square(numbers))
```

```python
2. Add 10 to every element in a list
```

```python
nums = [10, 20, 30]
add_result = list(map(lambda x: x + 10, nums))
print(add_result)

add_result2 = list(map(lambda x: x + 10, range(1,5)))
print(add_result2)
```

```python
# Conventional way of defining a function
def add_result(x): return x+10

nums = [10, 20, 30]
print(list(map(add_result, nums)))
```

```python
3. Add values in the lists Using Multiple iterables
Adding two arrays using map
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
add_result = list(map(lambda x, y: x + y, a, b))
print(add_result)

#      map(lambda arguments: expression, iterable)
```

```python
# Conventional way of defining a function
def add_result(x,y): return x+y

a = [1, 2, 3]
b = [4, 5, 6]
print(list(map(add_result, a, b)))
```

```python
4. Multiply All Elements by 3 (Using map)
Multiply a list with a number
```

```python
nums = [2, 4, 6, 8]
tripled = list(map(lambda x: x*3, nums))
print(tripled)   

t = list(map(lambda x: x*3, range(1,5)))
print(t)
```

```python
# Conventional way of defining a function
def tripled(x): return x*3

nums = [2, 4, 6, 8]
print(list(map(tripled, nums)))
```

### b. Filter Function: Filters elements based on a condition (True/False).

```python
Filter Syntax: filter(lambda arguments: condition, iterable)
```

```python
1. Filter out the even numbers from a list using a lambda function.
```

```python
nums = [10, 15, 20, 25, 30]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

```python
# Conventional function
# def evens(x):
#     if x%2 ==0:
#         return x 
def evens(x): return x if x%2 ==0 else None

nums = [10, 15, 20, 25, 30]
print(list(filter(evens, nums)))
print(list(filter(evens, range(1,7))))
```

```python
2. Filter out words starting with "A" using a lambda function.
```

```python
words = ["Apple", "Banana", "Avocado", "Mango", "Apricot"]
a_words = list(filter(lambda w: w.startswith("A"), words))
print(a_words)
```

```python
# Conventional function
def a_words(w): return w.startswith("A")

words = ["Apple", "Banana", "Avocado", "Mango", "Apricot"]
print(list(filter(a_words, words)))
```

```python
3. Filter out Positive Numbers Only (Using a lambda function)
```

```python
nums = [-5, 3, -1, 0, 7, -9]
positive = list(filter(lambda x: x > 0, nums))
print(positive)   # Expected output: [3, 7]
```

```python
# Conventional function
def positive(x): return x if x > 0 else None

nums = [-5, 3, -1, 0, 7, -9]
print(list(filter(positive, nums)))
```

### c. Reduce Function: Reduces an iterable to a single value.

Requires functools.reduce.

```python
Reduce Syntax: reduce(lambda arguments: expression, iterable)
```

```python
1. Write a program to find sum of numbers in a list using reduce
```

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]
total = reduce(lambda a, b: a + b, nums)
print(total)  
```

```python
# Conventional function
from functools import reduce

def total(a,b): return a+b

nums = [1, 2, 3, 4, 5]
print(reduce(total, nums))
```

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]
product = reduce(lambda a, b: a * b, nums)
print(product) 
```

```python
2. Write a program using reduce and lambda to compute the
factorial of a number n.
```

```python
from functools import reduce

n = 5
fact = reduce(lambda x, y: x * y, range(1, n+1))
print(f"Factorial of {n}: {fact}")
```

```python
3. Program to find largest number in a list using reduce
```

```python
from functools import reduce

numbers = [10, 45, 2, 99, 23]
maximum = reduce(lambda x, y: x if x > y else y, numbers)
print(maximum)
```

```python
Comparison of the map(), filter(), and reduce() functions
```

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

# Map → Squares
squares = list(map(lambda x: x**2, nums))
print("Squares:", squares)

# Filter → Even numbers
evens = list(filter(lambda x: x % 2 == 0, nums))
print("Evens:", evens)

# Reduce → Product
product = reduce(lambda a, b: a * b, nums)
print("Product:", product)
```

```python
Difference between map(), filter(), and reduce()
```

```python
filter()	Select elements based on condition
map()	Modify each element
reduce()	Combine all elements into one value
```
