### 1. Importing Built-in Modules

```python
Example 1: Importing entire module
```

```python
import math

print("Square root of 16 is:", math.sqrt(16))
print("Value of pi:", math.pi)
```

```python
Example 2: Giving alias name
```

```python
import math as mh

print("Square root of 16 is:", mh.sqrt(16))
print("Value of pi:", mh.pi)
```

```python
Example 3: Importing specific functions
```

```python
from math import sqrt

print("Square root of 25:", sqrt(25))
print("Value of pi:", pi)
```

```python
from math import sqrt, pi

print("Square root of 25:", sqrt(25))
print("Value of pi:", pi)
```

```python
Example 4: Importing with *
```

```python
from math import *

print("Square root of 25:", sqrt(25))
print("Value of pi:", pi)
```

```python
Advantages of Using Modules

a) Code Reusability – write once, use multiple times.

b) Better Organization – code is structured and easier to manage.

c) Namespace Management – avoids naming conflicts.

d) Faster Development – reusing built-in and user-defined modules
   saves time.
```

### 2. Standard Modules in Python

```python
Python provides many built-in standard modules. Three commonly
  used ones are:

1) math → mathematical functions

2) random → generating random numbers

3) datetime → working with date & time
```

```python
1. math Module

The math module provides mathematical functions.
```

```python
import math

print("Square root of 25:", math.sqrt(25))
print("Factorial of 5:", math.factorial(5))
print("Ceil of 4.2:", math.ceil(4.2))
print("Floor of 4.8:", math.floor(4.8))
print("Value of pi:", math.pi)
print("Power (2^5):", math.pow(2, 5))
```

```python
2. random Module

The random module is used for generating random numbers.
```

```python
import random

print("Random number between 1 and 10:", random.randint(1, 10))
print("Random float between 0 and 1:", random.random())
print("Random choice from list:", random.choice(['apple', 'banana', 'cherry']))
print("Shuffled list:", end=" ")
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers)
```

```python
3. datetime Module

The datetime module deals with date and time.
```

```python
import datetime

# Current date and time
now = datetime.datetime.now()
print("Current Date and Time:", now)

# Specific date
d = datetime.date(2025, 8, 21)
print("Custom Date:", d)

# Extract components
print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)
```
