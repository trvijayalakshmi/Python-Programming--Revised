### 1. Defining and Calling a Function

```python
A function is a block of code which only runs when it is called.

You can pass data, known as parameters, into a function.
```

```python
def greet_user(name):
    print("Hello,", name)

greet_user("Alice")
greet_user("Bob")
```

### 2. Parameters or Arguments?

```python
The terms parameter and argument can be used for the same thing: 
    information that are passed into a function.

From a function's perspective:

A parameter is the variable listed inside the parentheses in 
the function definition.

An argument is the value that is sent to the function when it 
is called.
```

### 3. Return values

```python
A function can return a value using the return statement.
```

```python
def add(a, b):
    return a + b

result = add(10, 20)
print("Sum:", result)
```

### 5. Programs on Functions

```python
1. Write a function called gcd that takes parameters a and b and
returns their greatest common divisor.
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Test the function
result = gcd(68,4)
print("GCD is ", result)
print("GCD of 64 and 4 is:", gcd(64, 4))   # Output: 4
print("GCD of 25 and 15 is:", gcd(25, 15)) # Output: 5
```

```python
2. Write a function to Print the Fibonacci sequence using while loop.
```

```python
def fibonacci_sequence(n):
    a, b = 0, 1
    count = 0
    while count < n:
        print(a, end=" ")
        a, b = b, a + b
        count += 1

# Driver code
terms = int(input("Enter the number of terms: "))
print("Fibonacci sequence:")
fibonacci_sequence(terms)
```

```python
3. Write a function to find factorial of a number.
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

print("Factorial of 5 is:", factorial(5))
```

```python
4. Write a function called palindrome that takes a string 
argument and returns True if it is a palindrome and False otherwise. 
```

```python
# function which return reverse of a string
def isPalindrome(s):
    return s == s[::-1]

s = "malayalam"
print(isPalindrome(s))
```

```python
def isPalindrome(st):
    for i in range(0, len(st)//2):
        if st[i] != st[len(st)-i-1]:
            return False
    return True
 
s = "malayalam"
print(isPalindrome(s))
if isPalindrome(s):
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

```python
5. Write a function to print all prime numbers in an interval
```

```python
def prime(x, y):
    for i in range(x, y):
        if i == 0 or i == 1:
            continue
        else:
            for j in range(2, int(i/2)+1):
                if i % j == 0:
                    break
            else:
                print(i,end=" ")

# Driver program
starting_range = 6
ending_range = 30
prime(starting_range, ending_range)
```

```python
7. Write a function to convert a decimal number to binary.
```

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"
    b = ""
    while n > 0:
        rem = n % 2
        b = str(rem) + b
        n = n // 2
    return b

num = 25
print("Binary:", decimal_to_binary(num))
```

```python
8. Program that computes harmonic sum of n.
```

```python
def harmonic_sum(n):
    total = 0.0
    for i in range(1, n + 1):
        total += 1 / i
    return total

# Example usage
n = int(input("Enter a number: "))
print(f"Harmonic sum of {n} = {harmonic_sum(n):.3f}")
```

```python
9. Write a function that returns a dictionary of squares of the 
numbers in a list.

Eg. of Function that takes a list as parameter and returns a dict.
```

```python
def sq(n):
    squares = {x: x*x for x in n}
    return squares

n = [3, 2, 5, 7]
result = sq(n)
print("The dict of squares for the given list ",result)
```

```python
10. Write a function to find the longest word in a string

Eg. of Function that takes a string as parameter and returns a string.
```

```python
def longest_word(s):
    words = s.split(" ")
    longest = max(words, key=len)
    return longest

s = "Python programming is easy to learn"
longest = longest_word(s)
print("Longest word:", longest)
```

```python
11. Function to invert a dictionary

Eg. of Function that takes a dictionary as parameter and returns 
a dictionary.
```

```python
def invert_dict(original):
    inverted = {value: key for key, value in original.items()}
    return inverted

original = {"a": 1, "b": 2, "c": 3}
print("The original dictionary is ",original)
inverted = invert_dict(original)
print("The inverted dictionary is ",inverted)
```
