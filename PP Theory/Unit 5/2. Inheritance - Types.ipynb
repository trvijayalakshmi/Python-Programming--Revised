### Python Inheritance: 

```python
Inheritance allows a class (child) to inherit properties and methods
from another class (parent).
```

```python
Uses of Inheritance:

a) Since a child class can inherit all the functionalities of the
parent's class, this allows code reusability.
b) Once a functionality is developed, you can simply inherit it. 
No need to reinvent the wheel.
c) Since you can also add your own functionalities in the child 
class, you can inherit only the useful functionalities and define 
other required features.
```

```python
Types of Inheritance:
    1. Single inheritance
    2. Multiple Inheritance
    3. Multilevel Inheritance
    4. Hierarchical Inheritance
```

```python
1. Single Inheritance

A child class inherits from one parent class.
```

```python
class Student:  # Patent class
    def college_details(self):
        print("The student belongs to MGIT")

class Branch(Student):   # Child class
    def section_details(self):
        print("He belongs to ECE.")

# s1 = Student()  # Try this Object creation with parent class
s1 = Branch()   #Object creation with child class
s1.college_details()   # use of parent method
s1.section_details()    # use of child method
```

```python
2. Multiple Inheritance

A child class inherits from more than one parent class.
```

```python
class A:      # Parent 1
    def method_A(self):
        print("Method from A")

class B:     # Parent 2
    def method_B(self):
        print("Method from B")

class C(A, B):   # Multiple Inheritance
    def add(self, a, b):        
        print("sum of a and b is ", a+b)

obj_c = C()    # Object creation from child class
obj_c.method_A()   # accessing method of parent class A
obj_c.method_B()   # accessing method of parent class B
obj_c.add(4,6)      # accessing method of Child class C
```

```python
Example of Multiple Inheritance:
```

```python
# Parent Class 1
class Teacher:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject

    def teach(self):
        print(f"{self.name} is teaching {self.subject}.")

# Parent Class 2, 
class Researcher:
    def __init__(self, field):
        self.field = field

    def research(self):
        print(f"She is researching in the field of {self.field}.")

# Derived Class (inherits from both Teacher and Researcher)
class Professor(Teacher, Researcher):
    def __init__(self, name, subject, field):
        Teacher.__init__(self, name, subject)      # Initialize Teacher part
        Researcher.__init__(self, field)  # Initialize Researcher part

    def display(self):
        print(f"Prof: {self.name}, Subject: {self.subject}, Field: {self.field}")

# Object creation
p = Professor("ECE", "Python Programming", "Deep learning")
p.display()
p.teach()
p.research()
```

```python
3. Multilevel Inheritance

A class is derived from another derived class (chain).
```

```python
class Grandparent:
    def show_gp(self):
        print("I am Grandparent")

class Parent(Grandparent):
    def show_p(self):
        print("I am Parent")

class Child(Parent):
    def show_c(self):
        print("I am Child")

c = Child()  # Object creation with child class
c.show_gp()
c.show_p()
c.show_c()
```

```python
Example of Multilevel Inheritance
```

```python
class Person:
    def __init__(self, name): 
        self.name = name

class Employee(Person):  # Employee (inherits from Person)
    def __init__(self, name, emp_id):
        Person.__init__(self, name)
        self.emp_id = emp_id  # additional attribute: emp_id
#     def employee_details(self):
#         print(f"Employee name is {self.name} with ID {self.emp_id}")

class Manager(Employee):   # Manager (inherits from Employee)
    def __init__(self, name, emp_id, dept):
        Employee.__init__(self, name, emp_id)
        self.dept = dept  # additional attribute: dept.

    def display(self):
        print(f"Name: {self.name}, ID: {self.emp_id}, Dept: {self.dept}")

m = Manager("Rajesh", 101, "ECE")
m.display()
# m.employee_details()
```

```python
When the object m is created, the Manager constructor runs.

It calls Employee.__init__(name, emp_id) → which runs Employee 
constructor.

The Employee constructor then calls Person.__init__(name) → 
which runs Person constructor.
```

```python
4. Hierarchical Inheritance

Multiple child classes inherit from the same parent class.
```

```python
class Parent:
    def greet(self):
        print("Calculate perimeter or area for given sides")

class Child1(Parent):
    def peri(self, a, b):
        d= 2*(a+b)
        print("Result of Child1 method ", d)

class Child2(Parent):
    def area(self, h, b):
        a = h * b
        print("Result of Child2 method ", a)

c1 = Child1()
c2 = Child2()
c1.greet()
c1.peri(4,3)
c2.greet()
c2.area(4,3)
```

```python
Examples of Hierarchical Inheritance:
```

```python
# Parent class
class Polygon:
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])
            
poly = Polygon(4)    
poly.inputSides()
poly.dispSides()
```

```python
A triangle is a polygon with 3 sides. So, we can create a class 
called Triangle which inherits from Polygon. This makes all the 
attributes of Polygon class available to the Triangle class.

We don't need to define them again (code reusability). Triangle 
can be defined as follows.
```

```python
# Child 1 class
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides
        # calculate the perimeter
        s = (a + b + c) 
        print('The perimeter of the triangle is %0.2f' %s)
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)
        
# Creating an instance of the Triangle class
t = Triangle()

# Prompting the user to enter the sides of the triangle
t.inputSides()

# Displaying the sides of the triangle
t.dispSides()

# Calculating and printing the area of the triangle
t.findArea()
```
