## 1. Python Classes and Objects

```python
Object: Everything is object and every object will have attributes 
and methods.

a) Attributes: Attributes are variables inside a class. They describe properties of
an object.

b) Methods: Methods are functions inside a class. They define
behavior of objects.
```

```python
1. Classes

A class is a blueprint for creating objects.
It defines attributes (variables) and methods (functions).
```

```python
class Student:   # Class creation
    name = "ECE"   # Attributes
    age = 20
    def student_details():   # Method
        print(f"My name is ECE and I am 20 years old.")
```

```python
2. Object creation and Accessing attribute

An object is an instance of a class.
Objects hold data and can call class methods.
```

```python
class Student:   # Class creation
    name = "ECE"   # Attributes
    age = 20
    def student_details(self):   # Method
        print("The grades of ECE are good")
#         print(f"My name is {name} and I am {age} years old.")
        
s1 = Student()   # Object creation
print(s1.name)  # Accessing attribute
print(s1.age)   # Accessing attribute
s1.student_details()       # Calling method
```

### The _ _init_ _() Function


```python
3. Constructor (__init__):

A constructor is a special method in Python classes.
It is automatically called when an object is created.
Defined using __init__.

A Class is like an object constructor, or a "blueprint" for 
creating objects.
```

```python
Note: 
The __init__() function is called automatically every time the class
is being used to create a new object.

The self parameter is a reference to the current instance of the
class, and is used to access variables that belong to the class.
```

```python
class Student:
    def __init__(self, name, age):  # constructor
        self.name = name      # Attribute
        self.age = age     # Attribute
    def student_details(self):   # Method
        print(f"My name is {self.name} and I am {self.age} years old.")

s1 = Student("Ravi", 19)   # Object creation
s2 = Student("Sanjay", 21)
print(s1.name, s1.age)      # Accessing attributes
print(s2.name, s2.age)
s1.student_details()    # calling method
```

```python
Another example of class, object, attributes, method
```

```python
class Calculator:
    def add(self, a, b): return a + b
    def sub(self, a, b): return a - b
    def mul(self, a, b): return a * b
    def div(self, a, b): return a / b if b != 0 else "Error: Division by zero"

calc = Calculator()
print(calc.add(10, 5))
print(calc.sub(10, 5))
print(calc.mul(10, 5))
print(calc.div(10, 0))
```

```python
Another example of class, object, method, attributes:
```

```python
# Defining a simple class "Person"
class Person:
    # Constructor to initialize Attributes
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

    # Define Methods
    def greet(self):
        print(f"Hello, my name is {self.name} and my roll no is {self.roll}.")

# Creating objects (instances) of the class "Person"
person1 = Person("Alice", 301)
person2 = Person("Bob", 254)

# Accessing attributes and methods of the objects
print(person1.name)  # Output: "Alice"
print(person2.roll)   # Output: 254
person1.greet()     
person2.greet()     
```

### The self Parameter
The self parameter is a reference to the current instance of the class, and is used to access variables that belongs to the class.

It does not have to be named self, you can call it whatever you like, but it has to be the first parameter of any function in the class:

```python
#Use the words mysillyobject and abc instead of self:

class Person:
    def __init__(mysillyobject, name, age):
        mysillyobject.name = name
        mysillyobject.age = age

    def myfunc(abc):
        print(f"Hello my name is {abc.name}, and I am {abc.age}")

p1 = Person("Sanjana", 13)
p1.myfunc()
```

```python
Constructors and types of constructors:

Constructors are generally used for instantiating an object.

The task of constructors is to initialize (assign values) to the
data members of the class when an object of the class is created.

In Python the __init__() method is called the constructor and is
always called when an object is created.
```

## Types of constructors : 

Default constructor: The default constructor is a simple constructor which doesn’t accept any arguments. Its definition has only one argument which is a reference to the instance being constructed.

Parameterized constructor: Constructor with parameters is known as parameterized constructor. The parameterized constructor takes its first argument as a reference to the instance being constructed known as self and the rest of the arguments are provided by the programmer.

```python
1. Default Constructor:
```

```python
class students:
 
    # default constructor
    def __init__(self):
        self.name = "ECE students"
    
    # a method for printing data members
    def display(self):
        print(self.name)
        
# creating object of the class
obj = students()
 
# calling the instance method using the object obj
obj.display()
```

```python
2. Parameterized Constructor:
```

```python
class Addition:
    answer = 0
     
    # parameterized constructor
    def __init__(self, f, s):
        self.f = f
        self.s = s
        
    def display(self):
        print("First number = " + str(self.f))
        print("Second number = " + str(self.s))
        print("Addition of two numbers = " + str(self.answer))
 
    def calculate(self):
        self.answer = self.f + self.s

# creating object of the class 
# this will invoke parameterized constructor
obj = Addition(1000, 2000)
# Perform calculation
obj.calculate() 
# display result
obj.display()
```

## Destructor

```python
# Python program to illustrate destructor
class students:
 
    # default constructor
    def __init__(self):
        self.geek = "ECE students"
    
    # a method for printing data members
    def display(self):
        print(self.geek)
        
# creating object of the class
obj = students()
del obj
```
