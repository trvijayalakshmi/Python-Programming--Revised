```python
A dictionary is a collection of key–value pairs.

Keys are unique and immutable (strings, numbers, tuples).

Values can be of any data type.
```

### 1. Creating Dictionaries

```python
dict1 = {"name": "Alice", "age": 25, "city": "New York"}
dict2 = dict([("name", "Alice"), ("age", 25), ("city", "New York")])
dict3 = dict(name = "Alice", age = 25, city = "New York")
dict4 = {}  # empty dictionary

print(dict1)
print(dict2)
print(dict3)
print(dict4)
print(len(dict1))
print(len(dict4))
```

### 2. Accessing and Changing Dictionary Elements: Using keys.

```python
# Example 1
person = {"name": "John", "age": 30, "country": "India"}

print(person["name"])      # John
print(person.get("age"))
print(person.get("age",'Not found'))   # 30
```

```python
Using a default value
```

```python
print(person.get("gender"))
print(person.get("gender", "Not Found"))  # Default value if key not found
```

```python
Difference between .get() vs [] 
If key is not found .get() gives None 
However if key is not found using [] gives Keyerror
```

```python
print(person.get('gender'))  # ✅ None
print(person['gender'])   # ❌ KeyError
```

```python
# Adding items to Dictionary
country_capitals = {
  "United States": "New York", 
  "Italy": "Naples" 
}

# add an item with "Germany" as key and "Berlin" as its value
country_capitals["Germany"] = "Berlin"
print(country_capitals)
```

```python
# Changing a value
country_capitals = {
  "United States": "New York", 
  "Italy": "Naples", 
  "England": "London"
}

# change the value of "Italy" key to "Rome"
country_capitals["Italy"] = "Rome"

print(country_capitals)
```

```python
# Change a single key
country_capitals = {
  "United States": "New York", 
  "Italy": "Naples", 
  "England": "London"
}

country_capitals['India'] = country_capitals.pop('Italy')
print(country_capitals)
```

### 4. Iterating Over Dictionaries

```python
Dictionaries can be iterated in different ways:
```

```python
1. Iterating over keys
```

```python
person = {"name": "Rahul", "age": 22, "city": "Delhi"}
for key in person.keys():
    print(key)
```

```python
2. Iterating over values
```

```python
for value in person.values():
    print(value)
```

```python
3. Iterating over key-value pairs
```

```python
for key, value in person.items():
    print(key, ":", value)
```

### 5. Dictionary Comprehension

```python
Like list comprehension, but with key: value pairs.
```

```python
1. Squares of numbers using dictionary comprehension
```

```python
# with comprehension
squares = {x: x*x for x in range(1, 11)}
print(squares)   # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
# without comprehension
squares = {}
for x in range(1,11):
    squares[x] = x*x
print(squares)
```

```python
2. Squares of even numbers using dictionary comprehension
```

```python
squares = {x: x*x for x in range(1, 10) if x%2==0}
print(squares) 
```

```python
3. Create a Dictionary of Odd/Even Labels for the numbers over
the range 1 to 15
```

```python
labels = {n: ("Even" if n % 2 == 0 else "Odd") for n in range(1, 16)}
print(labels)
```

```python
4. Write a Python program to count the frequency of each character
in a given string using a dictionary.
```

```python
# With comprehension
text = "banana"
freq = {ch: text.count(ch) for ch in text}
print(freq)  
```

```python
# Without comprehension
text = "banana"

freq = {}
for ch in text:
    freq[ch] = text.count(ch)

print("Character Count:", freq)
```

```python
5. Write a Python program to count the frequency of each word
in a given string using a dictionary.
```

```python
text = "apple banana apple orange banana apple"
words = text.split()

freq = {}
for i in words:
    freq[i] = words.count(i)

print("Word Frequency:", freq)
```

```python
6. Inverting a Dictionary
```

```python
# With comprehension
original = {"a": 1, "b": 2, "c": 3}
inverted = {value: key for key, value in original.items()}
print(inverted)
```

```python
# Without Comprehension
my_dict = {"a": 1, "b": 2, "c": 3}
print("Original Dict", my_dict)

inverted_dict = {}
for key, value in my_dict.items():
    inverted_dict[value] = key
    
print("Inverted Dict", inverted_dict)
```

```python
7. Remove Duplicate Values in a Dictionary
```

```python
data = {"a": 104, "b": 130, "a": 105, "d": 130}
print("Before Removing Duplicates:", data)
data2 = {}
for key, value in data.items():
    if value not in data2.values():
        data2[key] = value

print("After Removing Duplicates:", data2)
```

```python
8. Find the Student with Maximum marks
```

```python
x = {"Alice": 85, "Bob": 92, "Charlie": 89}
#For single topper
highest = max(x.values())
topper = max(x, key=x.get)
print(f"Topper is: {topper} with {highest} marks")
```

```python
To find all the toppers
```

```python
x = {"Alice": 85, "Bob": 92, "Charlie": 92}
#For multiple toppers
highest = max(x.values())
toppers = []
for key, value in x.items():
    if value == highest:
        toppers.append(key)

print(f"Toppers are: {toppers} with {highest} marks")
```

```python
# For multiple toppers Using list comprehension
x = {"Alice": 85, "Bob": 92, "Charlie": 92}
highest = max(x.values())
toppers = [key for key, value in x.items() if value == highest]
print(f"Toppers are: {toppers} with {highest} marks")
```

```python
9. Program to store names and marks of 'n' students in a 
dictionary and print the toppers with maximum marks.
```

```python
students = {}
x = {"Alice": 85, "Bob": 92, "Charlie": 92}

n = int(input('Enter the number of students'))
for i in range(n):
    name = input(f"Enter name of student {i+1}: ")
    marks = int(input(f"Enter marks of {name}: "))
    students[name] = marks

# Find the student with the highest marks
highest = max(students.values())

# Find all toppers using list comprehension   
toppers = [name for name, marks in students.items() if marks == highest]

print("\nStudents and their marks:", students)
print(f"Toppers: {toppers} with {highest} marks")
```
