### 2. Properties of Sets

```python
1. Unordered: The elements of a set do not have a fixed order.
```

```python
s = {"apple", "banana", "cherry"}
print(s)   # Order may vary each time
```

```python
2. Unique Elements/ No Duplicates : A set cannot contain 
duplicate values. If duplicates are added, Python automatically
removes them.
```

```python
s = {1, 2, 2, 3, 4, 4}
print(s)  
```

```python
3. No Indexing / Slicing: Since sets are unordered, you cannot 
access elements by index like in lists.
```

```python
s = {10, 20, 30}
# print(s[0])   ❌ Error
```

```python
4. Membership Testing: We can check if an element exists in a 
set using in.
```

```python
s = {1, 2, 3}
print(2 in s)   # True
print(5 in s)   # False
```

```python
5. Mutable (Changeable): A set itself is mutable → we can add 
or remove elements.
```

```python
s = {1, 2, 3}
s.add(4)
print(s)   # {1, 2, 3, 4}
```

```python
6. Immutable Elements: 
i) The elements inside a set must be immutable 
   (e.g., numbers, strings, tuples).
ii) Lists and dictionaries cannot be elements of a set.
```

```python
s = {1, "hello", (2, 3)}   # ✅ valid
# s = {[1, 2], 3}   ❌ ERROR (list is mutable)
```

### 3. Set Operations

```python
Sets Supports Mathematical Set Operations: 
Union ( ∪ ), Intersection ( ∩ ), Difference ( − ), 
Symmetric Difference.
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union (A ∪ B)
print("Union:", A | B)        
print("Union method:", A.union(B))

# Intersection (A ∩ B)
print("Intersection:", A & B)  
print("Intersection method:", A.intersection(B))

# Difference (A - B)
print("Difference:", A - B)    
print("Difference method:", A.difference(B))

# Symmetric Difference (A Δ B → elements in A or B but not both)
print("Symmetric Difference:", A ^ B)  
```

```python
In short, Sets are:

Unordered, no duplicates, no indexing, mutable,
immutable elements, supports set operations.
```

```python
5. Program to check if two sets are disjoint
```

```python
A = {1, 2, 3}
B = {4, 5, 6}

if A.isdisjoint(B):
    print("Sets are disjoint (no common elements)")
else:
    print("Sets are not disjoint")
```

### 5. Set Methods

```python
** Explain any four methods of sets with examples.
```

```python
1. Add Items
Once a set is created, you cannot change its items, but you can
add new items.

To add one item to a set use the add() method.
```

```python
thisset = {"apple", "banana", "cherry"}
thisset.add("orange")
print(thisset)
```

```python
2. Add Sets
To add items from another set into the current set, use the update() method.
```

```python
x = {"apple", "banana", "cherry"}
y = {"pineapple", "mango", "papaya"}

x.update(y)
print(x)
```

```python
3. Add Any Iterable
The object in the update() method does not have to be a set, 
it can be any iterable object (tuples, lists, dictionaries etc.).
```

```python
#Add elements of a list to a set:

x = {"apple", "banana", "cherry"}
y = ["kiwi", "orange"]

x.update(y)
print(x)
```

```python
4. Remove Item
To remove an item in a set, use the remove(), or the discard()
method.
```

```python
#Remove "banana" by using the remove() method:

x = {"apple", "banana", "cherry"}
x.remove("bana")
print(x)
```

```python
5. Discard Item
If the item to remove does not exist, remove() will raise an error.
```

```python
#Remove "banana" by using the discard() method:

x = {"apple", "banana", "cherry"}
x.discard("bana")
print(x)
```

```python
6. Pop
pop() method removes the last item. Remember that sets are 
unordered, so you will not know what item that gets removed.

The return value of the pop() method is the removed item.
```

```python
#Remove the last item by using the pop() method:

x = {"apple", "banana", "cherry", "e"}
y = x.pop()

print(y)
print(x)
```

```python
7. Clear the Set
Sets are unordered, so when using the pop() method, you do not 
know which item that gets removed.
```

```python
#The clear() method empties the set:

x = {"apple", "banana", "cherry"}
x.clear()
print(x)
```

```python
8. Delete the Set
The del keyword will delete the set completely
```

```python
x = {"apple", "banana", "cherry"}

del x
print(x)
```

## Programs on Sets

```python
1. Program to find unique elements from a list
```

```python
numbers = [1, 2, 2, 3, 4, 4, 5, 6, 1, 7]
unique_numbers = set(numbers)   # removes duplicates
print("Unique elements:", unique_numbers)
```

```python
2. Program to find common elements between two sets
```

```python
A = {"apple", "banana", "cherry"}
B = {"banana", "cherry", "mango"}

common = A & B   # Intersection
print("Common elements:", common)
```

```python
3. Program to remove duplicates from a string using sets
```

```python
text = "programming"
unique_chars = "".join(set(text))
print("Unique characters:", unique_chars)
```

```python
4. Program to count unique words in a sentence
```

```python
sentence = "Python is easy and Python is powerful"
words = sentence.split()
unique_words = set(words)
print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
```

```python
5. Program to find common elements and elements only in the first set
```

```python
# Create two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Finding common elements
common = set1.intersection(set2)

# Finding elements present only in the first set
only_in_first = set1.difference(set2)

# Displaying results
print("Set 1:", set1)
print("Set 2:", set2)
print("Common Elements:", common)
print("Elements only in Set 1:", only_in_first)
```
