```python
A tuple represents a sequence of objects separated by commas
and enclosed in parentheses. A tuple is an immutable object,
which means it cannot be changed, and we use it to represent
fixed collection of items.
```

### 1. Creating a Tuple

```python
# Empty tuple
t1 = ()
print(t1)   # ()

# Tuple with integers
t2 = (1, 2, 3, 4)
print(t2)   # (1, 2, 3, 4)

# Tuple with mixed data types
t3 = (10, "apple", 3.14, True)
print(t3)   # (10, 'apple', 3.14, True)
```

```python
1. Single Element Tuple
```

```python
t5 = (10,)     # single element tuple
print(t5)      # (10,)
print(type(t5))

t6 = (10)      # just an integer
print(t6)      # 10
print(type(t6))
```

```python
2. Nested Tuples: A tuple can contain other tuples.
```

```python
nested = (1, (2, 3), (4, 5, 6))
print(nested)        
print(nested[1][1])  # (accessing inside nested tuple)
```

```python
3. Tuple using Constructor
```

```python
list_data = [1, 2, 3, 4]
t7 = tuple(list_data)
print(t7)  

string_data = "Python"
t8 = tuple(string_data)
print(t8) 
```

```python
** Explain any four properties of tuples with examples.

Properties of Tuples-

1. Tuples are Ordered.
2. Immutable → once created, elements cannot be changed.
3. Allows duplicates.
4. Defined using ( ).
5. Heterogeneous: Can store different data types 
    (int, string, float, etc.).
6. Indexable and iterable.
```

```python
1. Tuples are Ordered

Tuples maintain the order of elements as they are inserted.
```

```python
t = (10, 20, 30)
print(t[0])   # Output: 10 (first element)
print(t[2])   # Output: 30 (last element)
```

```python
2. Tuples are Immutable

Once created, the elements of a tuple cannot be changed
(no modification, addition, or deletion).
```

```python
t = (1, 2, 3)
# t[0] = 5   # ❌ Error: 'tuple' object does not support item assignment
print(t)      # Output: (1, 2, 3)
```

```python
3. Tuples can contain Duplicate Elements

Unlike sets, tuples allow duplicate values.
```

```python
t = (1, 2, 2, 3, 1)
print(t)  # Output: (1, 2, 2, 3, 1)
```

```python
4. Tuples can contain Heterogeneous Data

A tuple can store elements of different data types
(integers, strings, floats, lists, etc.).
```

```python
t = (10, "Python", 3.14, [1, 2, 3])
print(t)  
```

### 2. Accessing Tuple Elements

```python
1. Accessing Tuple Elements by Index
```

```python
my_tuple = (10, 20, 30, 40, 50)

print(my_tuple[0])  
print(my_tuple[-1])   # Negative indexing
print(my_tuple[2]) 
```

```python
2. Accessing Tuple Elements Using a Loop
```

```python
x = ("apple", "banana", "cherry")
for i in x:
    print(i)
```

```python
3. Looping through indicies
```

```python
x = ("apple", "banana", "cherry")
for i in range(len(x)):
    print(x[i])
```

```python
4. Accessing a Range of Elements (Tuple Slicing)
```

```python
fruits = ("apple", "banana", "cherry", "mango", "grape")

print(fruits[1:4])   # From index 1 to 3 → ('banana', 'cherry', 'mango')
print(fruits[:3])    # From start to index 2 → ('apple', 'banana', 'cherry')
print(fruits[2:])
```

### 3. Tuple Methods

```python
Since tuples are immutable, they have very few built-in methods:

  a) count(x) → Returns number of times x appears.

  b) index(x) → Returns index of first occurrence of x.
```

```python
my_tuple = (10, 20, 30, 20, 40, 20)

print(my_tuple.count(20))   
print(my_tuple.index(30)) 
```

```python
To find all the indices:
```

```python
my_tuple = (10, 20, 30, 20, 40, 20)
indices = []
for i, j in enumerate(my_tuple):
    if j == 20:
        indices.append(i)
        
print(indices)
```

### 4. Operations with Tuples (not methods, but useful):

```python
numbers = (10, 5, 20, 15)

print(len(numbers))    # 4
print(min(numbers))    # 5
print(max(numbers))    # 20
print(sum(numbers))    # 50
print(sorted(numbers)) 
```

```python
1. Reverse a tuple
```

```python
fruits = ("apple", "banana", "cherry", "mango")

reversed_tuple = fruits[::-1]   # slicing with step -1
print("Reversed Tuple:", reversed_tuple)
```

```python
2. To concatenate two or more tuples
```

```python
# We can use + sign to concatenate (just like strings).

tuple_1 = ('a', 'b')
tuple_2 = (3, 4)
print(tuple_1 + tuple_2)
```

```python
3. Deleting an element is not possible in a Tuple
Complete Tuple can be deleted
```

```python
thistuple = ("apple", "banana", "cherry")
del thistuple
print(thistuple)
```

```python
4. Multiplying a tuple
```

```python
# multiplying a tuple by an integer produces a tuple containing the 
# original tuple repeated that many times. 

a = (1, 7, 9, 8)
print(a * 2)
```

```python
5. Add tuple to a tuple. 
```

```python
#Create a new tuple with the value "orange", and add that tuple:
x = ("apple", "banana", "cherry")
y = ("orange",)  # single element tuple should end with ,
x += y

print(x)
```

```python
6. Difference Between Tuples, Lists, and Sets in Python
```

```python
import sys
a_list = ['abc', 'xyz', 123, 231, 13.31, 0.1312]
a_tuple = ('abc', 'xyz', 123, 231, 13.31, 0.1312)
a_set = {'abc', 'xyz', 123, 231, 13.31, 0.1312}
print('The list size:', sys.getsizeof(a_list), 'bytes')
print('The tuple size:', sys.getsizeof(a_tuple), 'bytes')
print('The set size:', sys.getsizeof(a_set), 'bytes')
```

### 5. Programs using Tuples

```python
1. Write a Python program to create a tuple with numbers 
and print all the elements at odd indices.
```

```python
numbers = (10, 20, 30, 40, 50, 60, 70, 80)

print("Tuple:", numbers)
print("Elements at odd indices:")

# Accessing elements at odd indices using slicing
for i in range(1, len(numbers), 2):   # start from index 1, step = 2
    print(f"Index {i}: {numbers[i]}")
```

```python
2. Write a program to convert a list into a tuple and print
its length.
```

```python
# Creating a list
my_list = [5, 10, 15, 20, 25]

# Converting list to tuple
my_tuple = tuple(my_list)

# Printing tuple and its length
print("List:", my_list)
print("Tuple:", my_tuple)
print("Length of tuple:", len(my_tuple))
```
