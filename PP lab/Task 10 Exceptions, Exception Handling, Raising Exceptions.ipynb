## 1. Exceptions and Exception handling

### Theory for Exceptions:

```python
Exceptions are errors that occur during program execution and
disrupt the normal flow of the program.

Built-in Exceptions:

1. ZeroDivisionError → Divide by zero.
2. NameError → Variable not found.
3. ValueError → Wrong type of value (e.g. int("abc")).
4. IndexError → Index out of range.
5. KeyError → Key not found in dictionary.

6. TypeError → Invalid operation between different types.
7. FileNotFoundError → File does not exist.
8. ModuleNotFoundError → Module not found.

Errors:
1. SyntaxError →  Error not exception
2. IndentationError → Error not exception
```

### Theory for Exception handling:

```python
For Exception handling, python uses following keywords or statements

try
except
else
finally

To detect exceptions, write a code that might cause an error 
and enclose it in a try-except block. The try block contains the 
code that could potentially raise an exception.

When an exception is raised inside the try block, the program 
jumps to the corresponding except block. The except block contains
the code to handle the exception gracefully, allowing the program 
to continue running.

Try block: Lets you test a block of code for errors.
Except block: Lets you handle the error.
Else block: Lets you execute code when there is no error.
Finally block: Always runs (cleanup code), whether exception occurs
or not.
```

### 2. Raising Exceptions Manually

```python
Using raise inside try-except
How to create, raise and handle user defined exceptions?
```

```python
1. Raise an exception manually to check the age entered
```

```python
try:
    age = int(input("Enter your age: "))
    if age < 0:
        raise ValueError
    print("Your age is:", age)
except ValueError:
    print("Age cannot be negative")
```

```python
2. Raise multiple exceptions manually to check ValueError and ZeroDivisionError
```

```python
try:  
    a = int(input("Enter a : "))  
    b = int(input("Enter b : "))  
    if b == 0:  
        raise ArithmeticError
    else:  
        print("a/b = ",a/b) 
except ValueError:
    print("The values must be numbers")
except ArithmeticError:  
    print("The value of b can't be 0")
```
