```python
1. Write a function called gcd that takes parameters a and b and
returns their greatest common divisor.
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Test the function
result = gcd(68,4)
print("GCD is ", result)
print("GCD of 64 and 4 is:", gcd(64, 4))   # Output: 4
print("GCD of 25 and 15 is:", gcd(25, 15)) # Output: 5
```

```python
2. Write a function to Print the Fibonacci sequence using while loop.
```

```python
def fibonacci_sequence(n):
    a, b = 0, 1
    count = 0
    while count < n:
        print(a, end=" ")
        a, b = b, a + b
        count += 1

# Driver code
terms = int(input("Enter the number of terms: "))
print("Fibonacci sequence:")
fibonacci_sequence(terms)
```

```python
3. Write a function that returns a list of squares of the numbers
in a list.
```

```python
def sq(n):
    squares = [x*x for x in n]
    return squares

n = [3, 2, 5, 7]
result = sq(n)
print("The list of squares for the given list ",result)
```

```python
4. Write a function called palindrome that takes a string argument
and returns True if it is a palindrome and False otherwise. 
```

```python
# function which return reverse of a string
def isPalindrome(s):
    return s == s[::-1]

s = "malayalam"
print(isPalindrome(s))
```

```python
def isPalindrome(st):
    for i in range(0, int(len(st)//2)):
        if st[i] != st[len(st)-i-1]:
            return False
    return True
 
s = "malayalam"
print(isPalindrome(s))
if isPalindrome(s):
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

```python
5. Write a function to print all prime numbers in an interval
```

```python
def prime(x, y):
    for i in range(x, y):
        if i == 0 or i == 1:
            continue
        else:
            for j in range(2, int(i/2)+1):
                if i % j == 0:
                    break
            else:
                print(i,end=" ")

# Driver program
starting_range = 6
ending_range = 30
prime(starting_range, ending_range)
```

```python
def swap(x, y): 
    x,y = y,x
    return x, y
  
# Driver code 
a = 4
b = 9
print(f"Before swapping, the values are {a} and {b}")
c, d = swap(a, b) 
print(f"After swapping, the values are {c} and {d}")
```

```python
7. Write a function to convert a decimal number to binary.
```

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"
    b = ""
    while n > 0:
        rem = n % 2
        b = str(rem) + b
        n = n // 2
    return b

num = 25
print("Binary:", decimal_to_binary(num))
```

```python
8. Program that computes harmonic sum of n
```

```python
def harmonic_sum(n):
    total = 0.0
    for i in range(1, n + 1):
        total += 1 / i
    return total

# Example usage
n = int(input("Enter a number: "))
print(f"Harmonic sum of {n} = {harmonic_sum(n)}")
```

```python
9. Write a function that computes compound interest using default
values for "rate of interest" and "time"
```

```python
def cal_interest(principal, rate=5, time=3):
    amount = principal * (1 + rate/100) ** time
    interest = amount - principal
    return interest

interest = cal_interest(30000)
print("The compound interest is:", interest)
interest = cal_interest(30000, 10)
print("The compound interest is:", interest)
interest = cal_interest(30000, 10, 4)
print("The compound interest is:", interest)
```

```python
10. Write a function that takes variable length arguments to 
find sum of all the numbers with a default value.
```

```python
def add_numbers(*nums):
    total = 0
    for n in nums:
        total += n
    return total

print(add_numbers(10, 20))
print(add_numbers(1, 2, 3, 4, 5))
```

```python
def add_numbers(a, *nums):
    total = a
    for n in nums:
        total += n
    return total
a = 5
print(add_numbers(a, 10, 20))
print(add_numbers(a, 1, 2, 3, 4, 5))
```

```python
11. Write a function to take employee details using multiple keyword
arguments and a normal argument.
```

```python
def employee(name, **details):
    print("Name:", name)
    for key, value in details.items():
        print(key, ":", value)

employee("Kiran", salary=40000, department="ECE", city="Bangalore")
employee("Sanjay", salary=80000, department="IT", Institute="MGIT")
employee("Sanjay", department="IT", Institute="MGIT")
# employee()
```
