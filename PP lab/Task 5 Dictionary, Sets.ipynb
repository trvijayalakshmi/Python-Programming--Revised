### a) Dictionary is a Mapping Type

```python
1. Program to store names and marks of 5 students in a dictionary
and print the topper marks.
```

```python
students = {}

n = int(input('Enter the number of students'))
for i in range(n):
    name = input(f"Enter name of student {i+1}: ")
    marks = int(input(f"Enter marks of {name}: "))
    students[name] = marks

# Find the student with the highest marks
highest = max(students.values())

# Find all toppers using list comprehension   
toppers = [name for name, marks in students.items() if marks == highest]

print("\nStudents and their marks:", students)
print(f"Toppers: {toppers} with {highest} marks")
```

```python
Find the Student with Maximum Value and Highest score
```

```python
x = {"Alice": 85, "Bob": 92, "Charlie": 78}
highest = max(x.values())
topper = max(x, key=x.get)
print(f"Topper is: {topper} with {highest} marks")
```

```python
#For multiple toppers
x = {"Alice": 85, "Bob": 92, "Charlie": 92}

highest = max(x.values())
toppers = []
for key, value in x.items():
    if value == highest:
        toppers.append(key)

print(f"Toppers are: {toppers} with {highest} marks")
```

```python
2. Create a Dictionary of Squares of even numbers using dictionary
comprehension.
```

```python
squares = {x: x*x for x in range(1, 11) if x%2==0}
print(squares) 
```

```python
# without comprehension
sq = {}
for i in range(1,11):
    if i%2 == 0:
        sq[i] = i*i
print(sq)
```

```python
3. Inverting a Dictionary
```

```python
# Without Comprehension
data = {"a": 40, "b": 20, "c": 10, "d": 30}

print("Original Dict", data)
inverted_dict = {}
for key, value in data.items():
    inverted_dict[value] = key
    
print("Inverted Dict", inverted_dict)
```

```python
4. Remove Duplicate Values from a Dictionary
```

```python
data = {"a": 40, "a": 20, "c": 10, "d": 10}

unique_data = {}
for key, value in data.items():
    if value not in unique_data.values():
        unique_data[key] = value

print("After Removing Duplicates:", unique_data)
```

```python
# with dictionary comprehension
data = {"a": 40, "b": 20, "c": 10, "d": 30}

inverted = {value: key for key, value in data.items()}
print(inverted)
```

```python
5. Write a Python program to create a dictionary of characters
and their frequency in a given string using dictionary comprehension.
```

```python
text = "programming"

freq = {}
for ch in text:
    freq[ch] = text.count(ch)

print("Character Count:", freq)
```

```python
text = "programming"
freq = {ch: text.count(ch) for ch in text}
print("Character Count:", freq)
```

```python
6. Write a Python program to create a dictionary of words and their
frequency in a given sentence using dictionary comprehension.
```

```python
# Using Dictionary comprehension
text = "Python is easy and Python is powerful"
words = text.split()
freq = {i: words.count(i) for i in words}
print("Character Count:", freq)
```

```python
text = "Python is easy and Python is powerful"
words = text.split()

freq = {}
for i in words:
    freq[i] = words.count(i)

print("Character Count:", freq)
```

```python
sentence = "Python is easy and Python is powerful"
words = sentence.split()
unique_words = set(words)
print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
```

```python
9. Program to find common elements and elements only in the first set
```

```python
# Create two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Finding common elements
common = set1.intersection(set2)

# Finding elements present only in the first set
only_in_first = set1.difference(set2)

# Displaying results
print("Set 1:", set1)
print("Set 2:", set2)
print("Common Elements:", common)
print("Elements only in Set 1:", only_in_first)
```
