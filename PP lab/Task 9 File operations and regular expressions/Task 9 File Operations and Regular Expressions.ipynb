## Theory for File operations:

```python
Python provides built-in functions to create, read, write, and
append files using the open() function.

A file operation takes place in the following order:

1. Open a file
2. Read or write (perform operation)
3. Close the file

File Modes:
Mode	Description
"x" 	Creates a new file, error if file exists.
"r" 	Read (default). Opens file, error if file doesn’t exist.
"w" 	Write. Creates new file or overwrites existing file.
"a" 	Append. Adds data at the end of the file.
```

```python
1. Create a file in write mode and read the content from it
```

```python
f = open("sample.txt", "w")  # Overwrites if file exists
f.write("Hello, this is Python File Handling!\n")
f.write("Writing second line.\n")
f.close()

# Reading back
f = open("sample.txt", "r")
print(f.read())
f.close()
```

```python
2. Demonstrate seek() and tell() file methods 
```

```python
f = open("sample.txt", "r")
print("Initial position:", f.tell())

f.read(5)
print("After reading 5 chars:", f.tell())

f.seek(7)
print("The current position after seek:", f.tell())
f.close()
```

```python
3. Write a Python code to merge two given file contents into a 
third file.
```

```python
file1 = open("sample1.txt",'w')
file1.write("This is python programming lab")

file2 = open('sample2.txt','w') 
file2.write("\nThis content is copyrighted")

file1 = open("sample1.txt",'r')
file2 = open("sample2.txt",'r')
output_file = open("merged.txt",'w')

for line in file1:
    output_file.write(line)
for line in file2:
    output_file.write(line)

output_file = open("merged.txt",'r')
print(output_file.read())
print("Files merged successfully")
```

```python
4. Write a program that reads a file and displays the number of 
words, number of vowels, blank spaces, lower case letters and 
uppercase letters.
```

```python
file = open('sample1.txt','r')
num_words = 0
num_vowels = 0
num_spaces = 0
num_lower = 0
num_upper = 0

for line in file:
    num_words += len(line.split())
    for char in line:
        if char.lower() in 'aeiou':
            num_vowels = num_vowels + 1
        if char == ' ':
            num_spaces = num_spaces + 1
        if char.islower():
            num_lower = num_lower + 1
        if char.isupper():
            num_upper = num_upper + 1
        
print(f"Number of words: {num_words}")
print(f"Number of vowels: {num_vowels}")   
print(f"Number of spaces: {num_spaces}")
print(f"Number of lowercase letters: {num_lower}")
print(f"Number of uppercase letters: {num_upper}")
```

```python
5. Create a CSV file in write mode and fill it with details of students and then read the data from it.
```

```python
import csv
with open("students.csv", "w", newline="") as f:
    wr = csv.writer(f)
    wr.writerow(["Name", "Age", "Marks"])   # Header
    wr.writerow(["Archana", 20, 85])
    wr.writerow(["Raj", 22, 90])
    wr.writerow(["Charitha", 21, 78])
    
with open("students.csv", "r") as f:
    re = csv.reader(f)
    for row in re:
        print(row)
```

## b) Regular Expressions

### Theory for Regular expressions:

```python
Python has a built-in package called re Module, which can be used to work with Regular Expressions.
To use it, we need to import the module.

Core functions of re module:

a) match: Returns a Match object if there is a match else 
          returns None.
b) search: Returns a Match object if there is a match anywhere 
           in the string.
c) findall: Returns a list containing all matches.
d) split: Returns a list where the string has been split at  
          each match.
e) sub: Replaces one or many matches with a string

Useful Methods of re module:

a) group() → Returns the matched string.
b) start() → Start index of match.
c) end() → End index of match.
d) span() → Tuple of (start, end).
```

```python
6. Write a python code to read a phone number and email-id from the
user and validate it for correctness.
```

```python
import re
# Regular expressions for validation
phone_pattern = r'\b[6-9]\d{9}\b' 
email_pattern = r'\b[\w\.-]+@[\w\.-]+\.\w{2,}\b'

# Read inputs
phone = input("Enter phone number: ")
email = input("Enter email ID: ")

# Validate phone
if re.match(phone_pattern, phone):
    print("Valid Phone Number")
else:
    print("Invalid Phone Number")

# Validate email
if re.match(email_pattern, email):
    print("Valid Email ID")
else:
    print("Invalid Email ID")
```

```python
# Description of pattern of Phone no
\b: This is a boundary anchor.

[6-9] Matches exactly one digit between 6 and 9.

\d{9} \d → any digit (0–9)  {9} → exactly 9 digits.

So this part matches 9 digits after the first digit.
```

```python
# Description of pattern of Email
[\w\.-]+ (Username part)

[] → character set
\w → letters, digits, underscore (a-z, A-Z, 0-9, _)
. → dot
- → hyphen
+ → one or more characters
```
