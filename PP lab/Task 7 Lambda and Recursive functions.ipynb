## a) Lambda Functions

### Theory for Lambda functions:

```python
Normal function: Can have multiple statements and expressions. 
Defined using def with a function name.


Lambda Function: Only one expression allowed. 
Nameless one-line functions using the lambda keyword.
Often used with functions like map(), filter(), sorted()

Lambda Syntax: lambda arguments: expression
```

```python
1. Write a lambda function to return the maximum of two numbers.
```

```python
maximum = lambda a, b: a if a > b else b
print(maximum(12, 8))  
```

```python
# Conventional way of defining a function
def maximum(a,b): return a if a>b else b

print(maximum(12, 8)) 
```

```python
2. Write a lambda function to return the sum of two numbers.
```

```python
add = lambda a, b: a + b
print(add(3, 7))  
```

```python
# Conventional function
def add(a,b): 
    return a + b

print(add(3, 7))
```

Map Syntax:  map(lambda arguments: expression, iterable)

```python
3. Multiply all Elements by 3 (Using map)
```

```python
nums = [2, 4, 6, 8]
tripled = list(map(lambda x: x*3, nums))
print(tripled)   # Expected output: [6, 12, 18, 24]
```

```python
# Conventional way of defining a function
def tripled(x): return x*3

nums = [2, 4, 6, 8]
print(list(map(tripled, nums)))
```

```python
4. Create a list of squares for the given list of numbers using map
```

```python
numbers = [4, 2, 6, 3, 5]
squares = list(map(lambda x: x**2, numbers))
print(squares) 
```

```python
# Conventional way of defining a function
def squares(x): return x**2

print(list(map(squares, [4, 2, 6, 3, 5])))
```

Filter Syntax: filter(lambda arguments: condition, iterable)

```python
5. Filter out words starting with "A" using a lambda function.
```

```python
words = ["Apple", "Banana", "Avocado", "Mango", "Apricot"]
a_words = list(filter(lambda w: w.startswith("A"), words))
print(a_words)
```

```python
# Conventional function
def a_words(w): return w.startswith("A")

words = ["Apple", "Banana", "Avocado", "Mango", "Apricot"]
print(list(filter(a_words, words)))
```

```python
6. Filter out the even numbers from a list using a lambda function.
```

```python
nums = [10, 15, 20, 25, 30]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

```python
# Conventional function
def evens(x): return x if x%2 ==0 else None

nums = [10, 15, 20, 25, 30]
print(list(filter(evens, nums)))
print(list(filter(evens, range(1,7))))
```

Reduce Syntax: reduce(lambda arguments: expression, iterable)

```python
7. Write a program using reduce and lambda to compute the factorial
of a number n.
```

```python
from functools import reduce

n = 5
factorial = reduce(lambda x, y: x * y, range(1, n+1))
print("Factorial of", n, "=", factorial)
```

## b) Recursive Functions

### Theory for Recursive functions:

```python
Recursion is a process in which a function calls itself directly
or indirectly to solve a problem.

Every recursive function has:

i) Base Case → The condition that stops recursion.
ii) Recursive Case → Function calls itself with a smaller/simpler
    input.
```

```python
9. Find GCD (Greatest Common Divisor) using Recursion.
```

```python
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print(gcd(48, 18))  
```

```python
10. Write a recursive function to print factorial of a Number.
```

```python
def factorial(n):
    if n == 0 or n == 1:   # Base case
        return 1
    else:                  # Recursive case
        return n * factorial(n - 1)

print(factorial(5))  # Output: 120
```

```python
11. Write a recursive function to print fibonocci sequence.
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n_terms = 10
print("Fibonacci sequence:")
for i in range(n_terms):
    print(fibonacci(i))
```

```python
12. Write a recursive function to check whether the string is
Palindrome or not.
```

```python
def is_palindrome(s):
    if len(s)<=1: return True
    if s[0]!=s[-1]: return False
    return is_palindrome(s[1:-1])
print(is_palindrome("madam"))  
print(is_palindrome("racecar"))
print(is_palindrome("python")) 
```

```python
13. Write a recursive function to find sum of the digits of a
given number.
```

```python
def sum_of_digits(n):
    if n < 10:
        return n
    else:
        return n % 10 + sum_of_digits(n // 10)

print(sum_of_digits(12345))
```

```python
14. Write a recursive function to compute base raised to the power 
of exponent.
```

```python
def power(base, exponent):
    if exponent == 0:
        return 1
    else:
        return base * power(base, exponent - 1)

print(power(3, 5)) 
```

```python
15. Write a recursive function to convert decimal number to binary.
```

```python
def binary(n):
    if n > 1:
        binary(n//2)
    print(n % 2,end = '')
    
n = int(input("enter the decimal number"))
binary(n)
```
